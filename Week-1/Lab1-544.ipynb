```python
import pandas as pd
import plotnine as p9
from plotnine import *
import numpy as np
```

# Plot 1

Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

Bubble plot of Income and Life Expectancy with size of bubble by Population and fill by Four Regions.

```python
df = pd.read_csv("q1data.csv")

df = df.dropna(subset=['population', 'life_exp', 'income', 'four_regions'])
```

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'salmon', 'europe':'gold'}

(ggplot(df, aes(x='income',
y='life_exp',
size='population',
fill='four_regions',
label='country'))+
geom_text(aes(x=700, y=38, label='2010'), size=100, color='lightgray',ha='left',va='bottom')+
geom_point(color='black', stroke=0.3)+
labs(x="Income", y="Life Expectancy",fill="World Regions", size='Population')+
scale_x_continuous(limits = (450,128000),breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],trans='log',labels=[500, 1000, 2000, 4000, 8000, '16k', '32k', '64k'])+
scale_y_continuous(limits=(20,90),breaks=range(20,91, 10))+
scale_fill_manual(values=region_colors)+
scale_size_continuous(range=(0, 22))+
theme(panel_grid_major=element_line(color='lightgray'),
panel_grid_minor=element_blank(),
panel_background=element_blank())
)
```

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'salmon', 'europe':'gold'}

(ggplot(df, aes(x='income',
y='life_exp',
fill='four_regions',
label='country'))+
geom_text(aes(x=700, y=38, label='2010'), size=100, color='lightgray',ha='left',va='bottom')+
geom_tile(aes(width=0.2, height=0.2))+
labs(x="Income", y="Life Expectancy",fill="World Regions", size='Population')+
scale_x_continuous(limits = (450,128000),breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],trans='log',labels=[500, 1000, 2000, 4000, 8000, '16k', '32k', '64k'])+
scale_y_continuous(limits=(20,90),breaks=range(20,91, 10))+
scale_fill_manual(values=region_colors)+
scale_size_continuous(range=(0, 22))+
theme(panel_grid_major=element_line(color='lightgray'),
panel_grid_minor=element_blank(),
panel_background=element_blank())
)
```

This tile plot is very similar to the one that is found on Gapminder, however, the main difference is that this one does not include a size factor for population like the other one does. For this reason this plot is less informative that the original one, but it still does a good job in representing the relationship of income and life expenctancy while factoring by region.

# Plot 2

Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

Bubble Chart of Exports as a percentage of GDP and Imports as a percentage of GDP with bubble size by Energy usage and fill by Four Regions.

```python
df2 = pd.read_csv("q2data.csv")

df2['exports'] = pd.to_numeric(df2['exports'], errors='coerce')
df2['imports'] = pd.to_numeric(df2['imports'], errors='coerce')
df2['energy'] = pd.to_numeric(df2['energy'], errors='coerce')
df2 = df2.dropna(subset=['exports', 'imports', 'energy', 'four_regions'])
```

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'salmon', 'europe':'gold'}

(ggplot(df2, aes(x='exports',
y='imports',
size='energy',
fill='four_regions',
label='country'))+
geom_text(aes(x=20, y=150, label='1997'), size=100, color='lightgray',ha='left',va='bottom')+
geom_point(color='black', stroke=0.3)+
labs(x="Exports (\% of GDP)", y="Imports (\% of GDP)",fill="World Regions", size='Energy Use')+
scale_x_continuous(limits=(0,220),breaks=range(20,221, 20))+
scale_y_continuous(limits=(0,400),breaks=range(50,401, 50))+
scale_fill_manual(values=region_colors)+
scale_size_continuous(range=(0, 17))+
theme(panel_grid_major=element_line(color='lightgray'),
panel_grid_minor=element_blank(),
panel_background=element_blank())
)
```

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'salmon', 'europe':'gold'}

(ggplot(df2, aes(x='exports',
                 y='imports',
                 fill='four_regions',
                 weight='energy')) +
 geom_text(aes(x=20, y=150, label='1997'), size=100, color='lightgray', ha='left', va='bottom') +
 geom_bar(stat='identity', position='dodge', color='black', width=5) +
 labs(x="Exports (% of GDP)", y="Imports (% of GDP)", fill="World Regions", size='Energy Use') +
 scale_x_continuous(limits=(0, 220), breaks=range(20, 221, 20)) +
 scale_y_continuous(limits=(0, 400), breaks=range(50, 401, 50)) +
 scale_fill_manual(values=region_colors) +
 scale_size_continuous(range=(0, 17)) +
 theme(panel_grid_major=element_line(color='lightgray'),
       panel_grid_minor=element_blank(),
       panel_background=element_blank())
)
```

The pro of using a bar chart is that you can see each country along the X-axis. The cons of using this type of graph to display this information is that a lot of the observations are on top of others which crowds the higher density area of the axis. This makes it hard to see which region many of the countries from this area are in and it leads to the viewers of the plot not recieving the necessary information.

# Plot 3

Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

Bubble plot of Internet Users and GDP with bubble size of Income and fill by Four Regions.

```python
df3 = pd.read_csv("q3data.csv")

df3 = df3.dropna(subset=['four_regions'])
```

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'salmon', 'europe':'gold'}

(ggplot(df3, aes(x='internet_users',
y='gdp',
size='income',
fill='four_regions',
label='country'))+
geom_text(aes(x=10, y=2000, label='2001'), size=100, color='lightgray',ha='left',va='bottom')+
geom_point(color='black', stroke=0.3)+
labs(x="Individuals using the Internet", y="GDP/capita",fill="World Regions", size='Income')+
scale_x_continuous(limits=(0,90),breaks=range(10,91, 10))+
scale_y_continuous(limits = (0,200000),breaks=[200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000],trans='log',labels=[200, 500, 1000, 2000, 5000, '10k', '20k', '50k', '100k'])+
scale_fill_manual(values=region_colors)+
scale_size_continuous(range=(1, 14))+
theme(panel_grid_major=element_line(color='lightgray'),
panel_grid_minor=element_blank(),
panel_background=element_blank())
)
```

```python
region_colors = {'africa': 'deepskyblue', 'americas': 'lime', 'asia': 'salmon', 'europe': 'gold'}

(ggplot(df3, aes(x='internet_users', y='gdp', fill='four_regions')) +
 geom_bin2d() +
 labs(x="Individuals using the Internet (%)", y="GDP per Capita", fill="World Regions") +
 scale_fill_manual(values=region_colors) +
 scale_y_log10(limits=(200, 200000), 
                breaks=[200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000], 
                labels=[200, 500, 1000, 2000, 5000, '10k', '20k', '50k', '100k']) +
 theme(panel_grid_major=element_line(color='lightgray'),
       panel_grid_minor=element_blank(),
       panel_background=element_blank())
)
```

This plot is pretty similar to the Gapminder plot, but it leaves out a lot of the data points since many of the points are overlaying each other. This leads to a lot of the information not being conveyed to the viewer and the plot not fully showing what it's supposed to. This graph also doesn't represent the size of energy usage which is an additional factor that the original plot shows so, overall this plot is not a good as the original plot.
